```python
import sys
import yaml
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

from scraper.Parser import *
from scraper.mongo_handler import Mongo

from collections import OrderedDict
```

```python
plt.style.use('ggplot')
pd.set_option('display.max_colwidth', -1)
plt.rcParams['figure.figsize'] = (16.0, 8.0)
```

```python
stadt = 'berlin'

with open('scraper/config/{stadt}.yaml'.format(stadt=stadt)) as f:
    config = yaml.load(f)

mongo = Mongo(config['mongo'])
```

```python
languages = {}

for WG in 
.find({}, {'languages': 1}):
    if 'languages' not in WG:
        continue
        
    for lang in WG['languages']:
        if lang not in languages:
            languages[lang] = 0
        languages[lang]+= 1

languages = OrderedDict(sorted(languages.items(), key=lambda v: -v[1]))
if '' in languages: del languages['']
```

```python
y = languages.values()
x = np.arange(1, len(y) + 1, 1)
ticks = [language.capitalize() for language in languages.keys()]

plt.bar(x, y)
plt.xticks(x, ticks, size=14, rotation='vertical');
plt.title('Sprachen in {stadt} WGs'.format(stadt=stadt.capitalize()));
```
